```python
#3a

import sympy as sp
t = sp.symbols('t')
x = sp.Function('x')(t)
f = 2*t*(1 + x ** 2)
ode = sp.Eq(sp.Derivative(x, t), f)
solution = sp.dsolve(ode, x, ics={x.subs(t, 0): 1})
print("Solution to equation with given initial conditions:", solution.rhs)
```

```python
#3b

import matplotlib.pyplot as plt
import numpy as np
import math

numsteps = 8
stepwidth = 0.1
t = np.linspace(0, numsteps * stepwidth, numsteps + 1)
x = [1]


for i in range(numsteps):
    
    mk = 2*t[i]*(1+x[-1]**2)

    x.append(x[-1] + stepwidth * mk)
    
print("Predicted value of x(1):", x[-1])
print("Actual value of x(1):", np.tan(0.8**2+np.pi/4))
print("Error:", np.abs(np.tan(0.8**2+np.pi/4)-x[-1]))
erroreuler1 = np.abs(np.tan(0.8**2+np.pi/4)-x[-1])

plt.plot(t, x, label = "Approximate tan(t^2 + pi/4)")
plt.plot(t,np.tan(t**2+np.pi/4), label = "Exact tan(t^2 + pi/4)")
plt.title("tan(t^2 + pi/4) Approximation - Euler's Method, step size = 0.1")
plt.xlabel("Time t")
plt.ylabel("Function x(t)")
plt.legend()
plt.grid(True)
plt.show()
```
