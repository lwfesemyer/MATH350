```python
#2a

import sympy as sp
t = sp.symbols('t')
x = sp.Function('x')(t)
f = x
ode = sp.Eq(sp.Derivative(x, t), f)
solution = sp.dsolve(ode, x, ics={x.subs(t, 0): 1})
print("Solution to equation with given initial conditions:", solution.rhs)
```

```python
#2b

import matplotlib.pyplot as plt
import numpy as np

numsteps = 10
stepwidth = 0.1
t = np.linspace(0, numsteps * stepwidth, numsteps + 1)
x = [1]

for i in range(numsteps):
    
    mk = np.e**t[i]

    x.append(x[-1] + stepwidth * mk)
    
print("Predicted value of x(1):", x[-1])
print("Actual value of x(1):", np.e**1)
print("Error:", np.abs(np.e**1-x[-1]))
erroreuler1 = np.abs(np.e**1-x[-1])

plt.plot(t, x, label = "Approximate e^t")
plt.plot(t,np.e**t, label = "Exact e^t")
plt.title("e^t Approximation - Euler's Method, step size = 0.1")
plt.xlabel("Time t")
plt.ylabel("Function x(t)")
plt.legend()
plt.grid(True)
plt.show()
```
